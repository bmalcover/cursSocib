# Unsupervised Learning: Kmeans

## 1. Import libraries

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
```

## 2. Load the data.

I will create synthetic data: 

```python
X, y_true = make_blobs(n_samples=300, centers=3, cluster_std=0.60, random_state=0)

plt.scatter(X[:, 0], X[:, 1], s=50)
plt.title("Data")
plt.xlabel("X1")
plt.ylabel("X2")
plt.show()
```

## 3. K-Means

```python
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
labels = kmeans.labels_
centers = kmeans.cluster_centers_
```

## 4. Visualize the results

```python
plt.scatter(X[:, 0], X[:, 1], c=labels, s=50, cmap='viridis')

plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='X')
plt.title("Clusters founds by K-Means")
plt.xlabel("X1")
plt.ylabel("X2")
plt.show()
```

## Mixture of Gaussian 

```python
gmm = GaussianMixture(n_components=3, covariance_type='full', random_state=0)
gmm.fit(X)
labels = gmm.predict(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=labels, s=40, cmap='viridis')
plt.title('Clustering por GMM')
plt.show()
```

n_components=3
This is the number of Gaussians or clusters that the model will try to fit to the data. We can see there are 3 groups, so it makes sense to use n_components=3.

covariance_type='full'
Indicates the type of covariance matrix to be used for each Gaussian component.



## EXERCISE: 

Use the K-Means clustering algorithm to identify natural groupings in the Iris dataset, and compare the clustering results with the true species labels.

```python
from sklearn.datasets import load_iris
```

Load the data and feature names:

```python
iris = load_iris()
X = iris.data
y = iris.target
feature_names = iris.feature_names
```

Apply kmeans and select labels and clusters centers 

```python
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
labels = kmeans.labels_
centers = kmeans.cluster_centers_
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', marker='o', label='Cluster KMeans')
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, marker='X', label='Centroides')
plt.xlabel(feature_names[0])
plt.ylabel(feature_names[1])
plt.title('Clusters K-means with Iris dataset')
plt.colorbar(label='Cluster')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', marker='o', label='Cluster KMeans')
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, marker='X', label='Centroides')
plt.xlabel(feature_names[0])
plt.ylabel(feature_names[1])
plt.title('True labels with Iris dataset')
plt.colorbar(label='Cluster')
plt.legend()
plt.show()
```
